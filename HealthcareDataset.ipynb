```python
print(len(df))                # total rows
print(df.duplicated().sum())  # total duplicated rows
```

```python
# Drop duplicates
df_clean = df.drop_duplicates()

# Check the new size and duplicates
print("Total rows after cleaning:", len(df_clean))
print("Duplicate rows after cleaning:", df_clean.duplicated().sum())
```

```python
import pandas as pd

print(df_clean.info())      # check column types
print(df_clean.describe())  # summary of numeric columns
print(df_clean.isna().sum())  # check for missing values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age distribution
sns.histplot(df_clean['Age'], bins=20)
plt.title('Patients per Age')
plt.show()

# Medical Condition count
sns.countplot(y='Medical Condition', data=df_clean)
plt.title('Number of Patients in the particular disease')
plt.show()

# Billing amount by admission type
sns.boxplot(x='Admission Type', y='Billing Amount', data=df_clean)
plt.title('Average billing Amount')
plt.show()
```
